```
!pip install gdown --quiet

import gdown
import pandas as pd

url_prolongations = 'https://drive.google.com/uc?id=1n2CIsPZYOIqMfV7b3aUtp1efc9U96xVt'
url_financial = 'https://drive.google.com/uc?id=1r3grFMz_k-XuWhRNLOAZCoLDdVYg2C7X'

# Скачиваем файлы
print("Скачиваем файл 1...")
gdown.download(url_prolongations, 'prolongations.csv')

print("\nСкачиваем файл 2...")
gdown.download(url_financial, 'financial_data.csv')

print("\n Файлы успешно скачаны.")

# Загрузка данных в DataFrame
try:
    prolongations_df = pd.read_csv('prolongations.csv', sep=';', encoding='utf-8')
    financial_df = pd.read_csv('financial_data.csv', sep=';', encoding='utf-8', skipinitialspace=True)

    print(prolongations_df.head())
    print("\n")
    print(financial_df.head())

except Exception as e:
    print("\n Ошибка")
```
